```python
from PIL import Image
def fix_broken_pdf_covers(pdfs):
    for pi in pdfs:
        try: 
            Image.open(pi.cover_page.file)
            pi.cover_page = None
            pi.get_cover_page()
            print('ok', pi)
        except OSError:
            print('fixed', pi)

pdfs = PrintIssue.objects.filter(issue__issue_name__in=['2/2013'])
fix_broken_pdf_covers(pdfs)
```
