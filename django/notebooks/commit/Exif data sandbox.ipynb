```python
for im in ImageFile.objects.all():
    im.add_exif_from_file()
    im.save()
    print(im, im.exif_data)
    
```

```python
from django.utils import timezone
for im in ImageFile.objects.all():
    dt = im.exif_data.get('DateTimeOriginal')
    if dt: 
        time = timezone.datetime.strptime(dt, '%Y:%m:%d %H:%M:%S')
    else:
        time = '-'
    artist = im.exif_data.get('Artist', '-')
    description = im.exif_data.get('ImageDescription', '')
    print(f'{im.pk} {artist} {time}\n{description}')
    
    
    
    
```

```python
keys = {}
for im in ImageFile.objects.all():
    for key in im.exif_data.keys():
        if key in keys:
            keys[key] += 1
        else:
            keys[key] = 1
            
for key in sorted(keys.keys()):
    print(f'{key:<30}: {keys[key]}')
```
